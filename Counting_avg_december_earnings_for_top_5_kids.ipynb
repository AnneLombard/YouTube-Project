```python
pandas as pd

# List of new per-day files and corresponding channel names
files_and_channels = [
    ('/content/Like_per_day.csv', 'Like_Nastya'),
    ('/content/Diana_Kids_Show_per_day.csv', 'Kids_Diana_Show'),
    ('/content/Toys_and_Colors_per_day.csv', 'Toys_and_Colors'),
    ('/content/Vlad_and_Niki.csv', 'Vlad_and_Niki')
]

# Function to process each channel's December earnings
def process_channel(file_path, channel_name):
    # Load the daily data
    daily_data = pd.read_csv(file_path)
    daily_data['date'] = pd.to_datetime(daily_data['date'])

    # Filter December 1–11 data
    december_1_11_data = daily_data[
        (daily_data['date'] >= '2024-12-01') & (daily_data['date'] <= '2024-12-11')
    ]
    dec_1_11_earnings = december_1_11_data['mean_earning_dol'].sum()

    # Filter the monthly data for the first 11 months of 2024
    channel_monthly = top_kids_monthly[
        (top_kids_monthly['youtuber'] == channel_name) &
        (top_kids_monthly['date'] >= '2024-01-01') &
        (top_kids_monthly['date'] < '2024-12-01')
    ]

    # Calculate daily average earnings and average 20-day earnings
    channel_monthly['daily_average_earning'] = (
        channel_monthly['estimation_earning_numeric'] / channel_monthly['days_in_month']
    )
    average_20_day_earning = channel_monthly['daily_average_earning'].mean() * 20

    # Combine December 1–11 actual earnings with estimated remaining earnings
    total_december_earning = dec_1_11_earnings + average_20_day_earning

    # Update the top_kids_monthly dataframe
    top_kids_monthly.loc[
        (top_kids_monthly['youtuber'] == channel_name) &
        (top_kids_monthly['date'] == '2024-12-01'),
        'estimation_earning_numeric'
    ] = total_december_earning

    print(f"Updated December earnings for {channel_name}: {total_december_earning}")

# Process all files and channels
for file_path, channel_name in files_and_channels:
    process_channel(file_path, channel_name)

# Save the updated top_kids_monthly dataset
top_kids_monthly.to_csv('/content/updated_top_kids_monthly.csv', index=False)
```
