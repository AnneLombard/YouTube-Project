# The channels with rank 18-20 was the wrong ones now the goal would be to erase them and replace by other channels

```python
import pandas as pd
```

# Importing the df to modify

```python
df=pd.read_csv('top20new.csv')
```

```python
df
```

# Importing the top 150 list i retrieved with the Youtube API to take from there 6 columns related to my new 18,19,20 ranked creators to my dataset

```python
df_150=pd.read_csv("top150.csv")

# I need to search throughout the df so i set Pandas options to display all rows and columns
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

df_150[:100]
```

```python
df_150.columns

```

# Now that i found the 3 channels i need i filter the dataset to only show 3 rows i need

```python
# List of channel names to search for
channel_names_to_search = [
    "Fede Vigevani",
    "Alejo Igoa",
    "आचार्य प्रशान्त - Acharya Prashant"
]

# filter df_150 for the specified channel names
filtered_df_150 = df_150[df_150['Title'].isin(channel_names_to_search)]

filtered_df_150

```

# Now i need to assign to the channels the ranks i need according to the dataframe i need to enrich

```python
# order by Subscribers because rank is based on subscribers
filtered_df_150_sorted = filtered_df_150.sort_values(by="Subscribers", ascending=False)

# assign ranks 18, 19, 20
filtered_df_150_sorted['rank'] = [18, 19, 20]

filtered_df_150_sorted
```

# Now i see that my columns names are slightly different from the names in the df i need so i change them manually so i won't have join problems

```python
# rename columns
filtered_df_150_sorted.rename(columns={
    "Title": "title",
    "Country": "country",
    "Subscribers": "subscribers",
    "Video_Views": "video_views",
    "Uploads": "uploads",
    "Date_of_Creation": "created_date"
}, inplace=True)

filtered_df_150_sorted
```

# Perfect! But since the original df was already cleaned i need to bring the data format here to the same output

```python
# convert the 'created_date' column to datetime
filtered_df_150_sorted['created_date'] = pd.to_datetime(filtered_df_150_sorted['created_date']).dt.strftime('%Y-%m-%d')

filtered_df_150_sorted
```

# Now let's prepare the original dataset, we'll take out the lines with rank 18,19,20 so we can replace them with correct ones we prepared before

```python
# Remove rows with rank 18, 19, and 20
df_cleaned = df[~df['rank'].isin([18, 19, 20])]
df_cleaned
```

# Time to join

```python
# outer join

merged_df = pd.merge(
    df_cleaned,
    filtered_df_150_sorted,
    how="outer"
)

merged_df
```

# Great! But there are values i am not able to scrap so i would fill them in by hand.

```python
# update values for 'आचार्य प्रशान्त - Acharya Prashant'
merged_df.loc[merged_df['title'] == 'आचार्य प्रशान्त - Acharya Prashant', [
    'channel_link', 'channel_type', 'subscribers_last_30_days',
    'daily_views_average', 'lowest_monthly_income',
    'highest_monthly_income', 'lowest_yearly_income',
    'highest_yearly_income', 'views_last_30_days'
]] = [
    'https://www.youtube.com/@ShriPrashant', 'education', 600000,
    3001508, 20500, 327800, 245800, 3900000, 90045000
]

# update values for 'Fede Vigevani'
merged_df.loc[merged_df['title'] == 'Fede Vigevani', [
    'channel_link', 'channel_type', 'subscribers_last_30_days',
    'daily_views_average', 'lowest_monthly_income',
    'highest_monthly_income', 'lowest_yearly_income',
    'highest_yearly_income', 'views_last_30_days'
]] = [
    'https://www.youtube.com/@fedevigevani', 'entertainment', 1400000,
    22186381, 151400, 2400000, 1800000, 29100000, 665591000
]

# update values for 'Alejo Igoa'
merged_df.loc[merged_df['title'] == 'Alejo Igoa', [
    'channel_link', 'channel_type', 'subscribers_last_30_days',
    'daily_views_average', 'lowest_monthly_income',
    'highest_monthly_income', 'lowest_yearly_income',
    'highest_yearly_income', 'views_last_30_days'
]] = [
    'https://www.youtube.com/@AlejoIgoa', 'entertainment', 2900000,
    15693823, 107100, 1700000, 1300000, 20600000, 470815000
]


merged_df
```

# Sort by rank to check if the nb of subscribers is coherent and that would show that there is no errors in ranks too

```python
# Sort the merged DataFrame by the 'rank' column
merged_df_sorted = merged_df.sort_values(by="rank", ascending=True)

# Display the sorted DataFrame
merged_df_sorted
```

# Looks good, time to export

```python

merged_df_sorted.to_csv("Top20_2024.csv", index=False)
```
