Scrapping du site : https://us.youtubers.me/mrbeast/youtube-estimated-earnings

```python
import requests
from bs4 import BeautifulSoup

url = " https://us.youtubers.me/mrbeast/youtube-estimated-earnings"

response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
date = soup.find_all("tr")
date
```

```python
date[1].text.split("$ ")
```

```python
data = {"date":[],"estimation_earning":[]}

for i in date[1:]:
    print(i.text.split("$ "))
    data["date"].append(i.text.split("$ ")[0])
    data["estimation_earning"].append(i.text.split("$ ")[1])
```

```python
import pandas as pd
df = pd.DataFrame(data)

def parse_earnings(value):
    if value.endswith('K'):  # Si la valeur se termine par 'K', convertir en milliers
        return float(value[:-1]) * 1_000
    elif value.endswith('M'):  # Si la valeur se termine par 'M', convertir en millions
        return float(value[:-1]) * 1_000_000
    return float(value)  # Sinon, convertir directement

df["estimation_earning_numeric"] = df["estimation_earning"].apply(parse_earnings)
df["date"] = pd.to_datetime(df["date"], format='%B %Y')
df
```

```python
# Exporter le DataFrame dans un fichier CSV
df.to_csv("mrbeast.csv", index=False)  # index=False pour exclure l'index du DataFrame dans le fichier CSV
```
