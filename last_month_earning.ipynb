```python
import pandas as pd

# Convertir la colonne 'date' en format datetime
df["date"] = pd.to_datetime(df["date"])

# Fonction pour convertir les colonnes en float (avec gestion des virgules et des suffixes)
def convert_to_float(value):
    if isinstance(value, str):
        value = value.replace(",", "")  # Supprimer les virgules
        multiplier = 1
        if "K" in value:  # Convertir les valeurs avec "K" (milliers)
            multiplier = 1_000
            value = value.replace("K", "")
        elif "M" in value:  # Convertir les valeurs avec "M" (millions)
            multiplier = 1_000_000
            value = value.replace("M", "")
        elif "B" in value:  # Convertir les valeurs avec "B" (milliards)
            multiplier = 1_000_000_000
            value = value.replace("B", "")
        return float(value) * multiplier
    return float(value)

# Appliquer les conversions sur les colonnes concernes
df["subs"] = df["subs"].apply(convert_to_float)
df["views"] = df["views"].apply(convert_to_float)

# Afficher le DataFrame aprs conversion
print(df.dtypes)
print(df)
```

```python
#Appliquer sur df1
# Fonction pour convertir les estimations en float (prendre la moyenne de la plage)
def convert_estimation(value):
    if isinstance(value, str):
        value = value.replace("$", "").replace(",", "")  # Supprimer les symboles et les virgules
        if " - " in value:  # Grer les plages de valeurs
            low, high = map(float, value.split(" - "))
            return (low + high) / 2  # Retourner la moyenne
        return float(value)  # Retourner la valeur unique si pas de plage
    return float(value)
```
