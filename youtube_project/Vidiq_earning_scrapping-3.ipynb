```python
import pandas as pd

# Convertir la colonne 'date' en format datetime
df["date"] = pd.to_datetime(df["date"])

# Fonction pour convertir les colonnes en float (avec gestion des virgules et des suffixes)
def convert_to_float(value):
    if isinstance(value, str):
        value = value.replace(",", "")  # Supprimer les virgules
        multiplier = 1
        if "K" in value:  # Convertir les valeurs avec "K" (milliers)
            multiplier = 1_000
            value = value.replace("K", "")
        elif "M" in value:  # Convertir les valeurs avec "M" (millions)
            multiplier = 1_000_000
            value = value.replace("M", "")
        elif "B" in value:  # Convertir les valeurs avec "B" (milliards)
            multiplier = 1_000_000_000
            value = value.replace("B", "")
        return float(value) * multiplier
    return float(value)

# Appliquer les conversions sur les colonnes concernées
df["subs"] = df["subs"].apply(convert_to_float)
df["views"] = df["views"].apply(convert_to_float)

# Afficher le DataFrame après conversion
print(df.dtypes)
print(df)
```

```python
# Diviser les revenus estimés en deux colonnes : min_earning et max_earning
df[["min_earning_dol", "max_earning_dol"]] = df["estimated_earning"].str.split(" - ", expand=True)

# Supprimer les symboles "$" et convertir en float
df["min_earning_dol"] = df["min_earning_dol"].str.replace("$", "").str.replace(",", "").astype(float)
df["max_earning_dol"] = df["max_earning_dol"].str.replace("$", "").str.replace(",", "").astype(float)
# Ajouter une colonne pour la moyenne
df["mean_earning_dol"] = (df["min_earning_dol"] + df["max_earning_dol"]) / 2
```
