```python
import pandas as pd
# objective : merge of 2 databases, YouTube Top 20 2023 and 2024 in order to make comparisons and graphs for project. Merge type: union all as no key or common data
```

# **Creation dataframe top 20 2023**

```python
df_2023 = pd.read_csv("/content/top_1000_stats_2023.csv")
df_2023.info()
```

```python
# select 20 first rows
df_2023 = df_2023.iloc[0:20, :]
df_2023.info()
```

# **Data cleaning**

```python
# select columns to remove: "category","video_views_rank","country_rank", "channel_type_rank", "gross_tertiary_education_enrollment", "population", "unemployment_rate", "urban_population"

df_2023_clean = df_2023.drop(["category","video_views_rank","country_rank", "channel_type_rank", "gross_tertiary_education_enrollment", "population", "unemployment_rate", "urban_population"], axis= 1)
df_2023_clean.info()
```

```python
# remove "rank" column
# df_2023_clean = df_2023_clean.drop(["rank"], axis = 1)
# df_2023_clean.info()
```

```python
# rename columns "video_views_for_the_last_30_days", "lowest_monthly_earnings", "highest_monthly_earnings", "lowest_yearly_earnings", "highest_yearly_earnings", "subscribers_for_last_30_days", "created_year"
df_2023_clean.rename(columns={'video_views_for_the_last_30_days': 'views_last_30_days', 'subscribers_for_last_30_days': 'subscribers_last_30_days', "created_year": "year"}, inplace=True)
df_2023_clean.columns = df_2023_clean.columns.str.replace("earnings", "income")
df_2023_clean.info()
```

```python
# change "year" type int
df_2023_clean["year"] = df_2023_clean["year"].astype(int)
df_2023_clean.info()
```

```python
# reorganize columns
df_2023_clean.insert(0, "rank", df_2023_clean.pop("rank"))
df_2023_clean.insert(1, "year", df_2023_clean.pop("year"))
df_2023_clean.insert(2, "country", df_2023_clean.pop("country"))
df_2023_clean.insert(7, "subscribers_last_30_days", df_2023_clean.pop("subscribers_last_30_days"))
df_2023_clean.info()
```

```python
# insert column "data_year"
df_2023_clean.insert(14, "data_year", 2023)
df_2023_clean
```

# **Creation dataframe top 20 2024**

```python
df_2024 = pd.read_csv("/content/Top20_2024.csv")
df_2024.info()
```

```python
# drop columns not into df_2023_clean
df_2024_clean = df_2024.drop(["channel_link","daily_views_average"], axis= 1)
df_2024_clean
```

```python
# keep year in "created_date"
df_2024_clean["created_date"] = df_2024_clean["created_date"].str[-4:]
# change data type "created_date" to int
df_2024_clean["created_date"] = df_2024_clean["created_date"].astype(int)
# rename column "created_date"
df_2024_clean = df_2024_clean.rename(columns = {"created_date" : "year"})
#df_2024_clean.info()
```

```python
# reorganize columns
print(df_2023_clean.columns)
df_2024_clean.columns
```

```python
df_2024_clean.insert(0, "year", df_2024_clean.pop("year"))
df_2024_clean.insert(1, "country", df_2024_clean.pop("country"))
df_2024_clean.columns
```

```python
df_2024_clean
```

```python
# remove non latin characters
df_2024_clean["title"] = df_2024_clean["title"].str.encode("ascii","ignore").str.decode("ascii")
```

```python
# replace NaN
df_2024_clean = df_2024_clean.fillna(0)
# change "subscribers_last_30_days" data type to int
df_2024_clean["subscribers_last_30_days"] = df_2024_clean["subscribers_last_30_days"].astype(int)
df_2024_clean.info()
```

```python
# insert "data_year" column
df_2024_clean.insert(13, "data_year", 2024)
df_2024_clean
```

# Join tables

```python
# union all join
df_merged = pd.concat([df_2023_clean, df_2024_clean], ignore_index= True)
print(df_merged.info())
df_merged
```
